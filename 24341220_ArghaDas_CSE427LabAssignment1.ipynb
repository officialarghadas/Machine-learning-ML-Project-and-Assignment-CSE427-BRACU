```python
import numpy as np

# Task 1: Create two linear arrays a and b with 20 random elements (1 to 100)
a = np.random.randint(1, 101, 20)
b = np.random.randint(1, 101, 20)
print("Linear array a:", a)
print("Linear array b:", b)

# Task 2: Reshape them into 5x4 matrices
a = a.reshape(5, 4)
b = b.reshape(5, 4)
print("\nMatrix a:\n", a)
print("\nMatrix b:\n", b)

# Task 3: Perform scalar multiplication
a_scaled = a * 2
b_scaled = b * 3
print("\nScalar multiplication of a (x2):\n", a_scaled)
print("\nScalar multiplication of b (x3):\n", b_scaled)

# Task 4: Modify matrix b and perform matrix multiplication with a to store in c
b_modified = b.T  # Transposing b to make it compatible for matrix multiplication
c = np.dot(a, b_modified)
print("\nMatrix multiplication (a * b_modified) stored in c:\n", c)

# Task 5: Print only column 2 and 3 of matrix c using slicing
print("\nColumns 2 and 3 of matrix c:\n", c[:, 1:3])

# Task 6: Find the largest element in c and its index
max_element = np.max(c)
max_index = np.unravel_index(np.argmax(c), c.shape)
print("\nLargest element in c:", max_element, "at index", max_index)

# Task 7: Convert c into a 1D array without modifying the original array
c_flattened = c.flatten()
print("\nFlattened matrix c:\n", c_flattened)
```

```python
import numpy as np
import pandas as pd

# Task 1: Read the CSV file as DataFrame
df = pd.read_csv('/content/subject_scores.csv')

# Task 2: Show the first few rows of the DataFrame
print("\nFirst few rows of the dataset:\n", df.head())

# Task 3: Show summarized details of the dataset
print("\nDataset summary:\n", df.info())

# Task 4: Show the number of null values in each column
print("\nNull values in each column:\n", df.isnull().sum())

# Task 5: Replace null values with the mean of the respective column
df.fillna(df.mean(), inplace=True)
print("\nNull values replaced with column mean.")

# Task 6: Show only the Math scores of all students
if 'Math' in df.columns:
    print("\nMath scores:\n", df['Math'])
else:
    print("\n'Math' column not found in dataset.")
```

```python
#1. Generate a scatter plot for temperature_dhaka vs humidity_dhaka using (*) as the marker in red.
import matplotlib.pyplot as plt

# Data
temperature_dhaka = [25, 34, 21, 45, 28, 6, 43, 18, 7, 2]
humidity_dhaka = [28, 25, 29, 20, 26, 50, 19, 29, 52, 55]


# Scatter plot
plt.scatter(temperature_dhaka, humidity_dhaka, color='red', marker='*')
plt.title('Temperature vs Humidity')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Humidity (%)')
plt.show()
```

```python
#2. Create a line plot for study_hours vs marks with figure size (12,8).

# Data
study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]

# Line plot
plt.figure(figsize=(12, 8))
plt.plot(study_hours, marks, marker='o', linestyle='-', color='blue')
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

```python
#3. Draw a horizontal bar graph showing the scores in each subject.

# Data
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
scores = [89, 90, 45, 78, 99]

# Horizontal bar graph
plt.barh(subjects, scores, color='green')
plt.title('Scores in Subjects')
plt.xlabel('Scores')
plt.show()
```

```python
#4. Draw a vertical bar graph where each subject score is visualized in different colors.

# Vertical bar graph
plt.bar(subjects, scores, color=['red', 'blue', 'green', 'purple', 'orange'])
plt.title('Scores in Subjects')
plt.ylabel('Scores')
plt.show()
```

```python
#5. Show all these graphs in one figure using the subplot() function.


# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Scatter plot
axes[0, 0].scatter(temperature_dhaka, humidity_dhaka, color='red', marker='*')
axes[0, 0].set_title('Temperature vs Humidity')
axes[0, 0].set_xlabel('Temperature (Â°C)')
axes[0, 0].set_ylabel('Humidity (%)')

# Line plot
axes[0, 1].plot(study_hours, marks, marker='o', linestyle='-', color='blue')
axes[0, 1].set_title('Study Hours vs Marks')
axes[0, 1].set_xlabel('Study Hours')
axes[0, 1].set_ylabel('Marks')

# Horizontal bar graph
axes[1, 0].barh(subjects, scores, color='green')
axes[1, 0].set_title('Scores in Subjects')
axes[1, 0].set_xlabel('Scores')

# Vertical bar graph
axes[1, 1].bar(subjects, scores, color=['red', 'blue', 'green', 'purple', 'orange'])
axes[1, 1].set_title('Scores in Subjects')
axes[1, 1].set_ylabel('Scores')

# Adjust layout
plt.tight_layout()
plt.show()
```
